```python
import pandas as pd
from sklearn.externals import joblib
from tkinter import *
import numpy as np
data3=pd.read_csv("C:\\Users\\KARAN\\Desktop\\machine learning\\a movie rec. system\\data3.csv")
train=pd.read_csv("C:\\Users\\KARAN\\Desktop\\machine learning\\a movie rec. system\\train.csv")
model=joblib.load("C:\\Users\\KARAN\\Desktop\\machine learning\\a movie rec. system\\model")
```

```python
root=Tk()
root.title("Movie Recommendation System")
#root.attributes('-fullscreen', True)
#root.geometry('400x100')
root.state('zoomed')
f3=Frame(root,bg='light blue')
f3.pack(expand=TRUE,fill=BOTH)
f1=Frame(root,bg='Cyan')
f1.pack(expand=TRUE,fill=BOTH,padx=5,pady=5)

l1=Label(f3,text='Movies Recomendation System \n \n ',bg='light blue',font=100,pady=40)
l1.pack()


k1=Label(f1,bg='Cyan',text='Enter any one of the following genres or genres of your choice seperated by a ","("comma") -\n')
k1.pack()

k4=Label(f1,bg='Cyan',fg='purple',text='Given below are a few suggestions -\n\n')
k4.pack()

f2=Frame(root,bg='Cyan')
f2.pack(expand=TRUE,fill=BOTH)
l=list(train.columns)

for i in range(7):
    #print(end="\n or \n")
    l2=Label(f1,bg='Cyan',text=l[np.random.randint(1,200)]+'\n\n')
    l2.pack()

f5=Frame(root,bg='light blue')
f5.pack(expand=True,fill=BOTH)
e1=Entry(f5)
e1.pack(pady=15)
#root.mainloop()
def submit():
    global l2
    l2.destroy()
    try:
        string=e1.get()
        test=train[data3['Genres'].str.find(string)!=-1]
        ar=model.predict(test.iloc[:,1:-1])
        ar=pd.DataFrame(ar)
        ar.drop_duplicates(inplace=True)
        e1.delete(0,END)
        #print("\n\nYour Recommendations are -")
        try:
            if(len(ar)<15):
                root1=Tk()
                f0=Frame(root1)
                f0.pack()
                l3=Label(f0,text=ar.iloc[:,:].values,bg='Cyan',font=100)
                l3.pack()
                root1.mainloop()
            else:
                root1=Tk()
                f0=Frame(root1)
                f0.pack()    
                l3=Label(f0,text=ar.iloc[0:15,:].values,bg='Cyan',font=100)
                l3.pack()                
                root1.mainloop()
        except:
            f=0
    except:
        l2=Label(f1,bg='Cyan',text="Sorry the genre is not present please try again. Do enter the genres in an alphabetical order.")
        l2.pack()

k3=Label(f1,bg='Cyan',fg='red',text='Or any other Genre(or Genres) written like the suggestions given above (in an alphabetical order) \n')
k3.pack()
        
k2=Label(f1,bg='Cyan',fg='red',text="*Enter the genres in an alphabetical order*\n\n")
k2.pack()
b1=Button(f5,text='Submit',bd=10,command=submit)
b1.pack(pady=15)

root.mainloop()
```
